```python
import os

val_path = 'val_images'

for dirc in os.listdir('prep_images'):
    dir_path = os.path.join(val_path, dirc)
    if not os.path.exists(dir_path):
        os.makedirs(dir_path)
        imgs = os.listdir(os.path.join('testing', dirc))
        for i in range(0, 20):
            os.rename(os.path.join('testing', dirc, imgs[i]), os.path.join(dir_path, imgs[i]))
        print('Validation set created.')
```
